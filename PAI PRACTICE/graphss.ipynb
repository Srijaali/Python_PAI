MATPLOT LIB

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.title("Matplotlib Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]
c = ['red','orange','green']
plt.bar(categories, values, color=c)
plt.title("Matplotlib Bar Plot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 3, 6, 9, 2]
y2 = [4,5,6,2,1]

plt.scatter(x, y1, color='green',label='x')
plt.scatter(x,y2,color='blue',label='y')
plt.title("Matplotlib Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.title("Matplotlib Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data = [7, 8, 8, 5, 6, 9, 10, 8, 6, 5]

plt.boxplot(data)
plt.title("Matplotlib Box Plot")
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
ex = [0.2,0.1,0,0]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,explode=ex)
plt.title("Matplotlib Pie Chart")
plt.show()
```

SEABORN

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({'x': x, 'y': y})
sns.lineplot(x='x', y='y', data=data)
plt.title("Seaborn Line Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Customizing colors
bar_colors = ['red', 'blue', 'green', 'orange']

# Creating the bar plot
sns.barplot(x=categories, y=values, palette=bar_colors)

# Adding a title
plt.title("Seaborn Bar Plot")

# Displaying the plot
plt.show()

#if dataframe:
import pandas as pd
df = pd.DataFrame({'Category': categories, 'Value': values})
sns.barplot(x='Category', y='Value', data=df, color='cyan')
```
