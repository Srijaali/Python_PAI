```python
#scatter plot
x = ['mon' , 'tue','wed','thur','fri']
y = [24,34,21,12,27]
plt.xlabel('days')
plt.ylabel('temp')
plt.title('temp each day',fontsize=24)
sizes = [220,225,320,325,420]
c = ['red','orange','yellow','green' , 'blue']
plt.scatter(x,y,s=sizes,color=c,marker='^')
plt.show()
```

```python
#pie-chart
import matplotlib.pyplot as plt
x = [20,25,50,75]
y = ['c','c++','java','python']
ex = [0.2,0,0,0]
c = ['red','orange','yellow','blue']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.01f%%')
plt.show()
```

```python
##stack
import matplotlib.pyplot as plt
x= [1,2,3,4]
a=[1,2,3,5]
a2=[1,3,5,7]
a3=[4,6,3,2]
l = ['a1','a2','a3']
plt.stackplot(x,a,a2,a3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
#box plot
import matplotlib.pyplot as plt
x=[20,40,60,80]
x1=[75,55,61,45]
y = [x,x1]
plt.boxplot(y,vert=True,widths=0.4,labels=['python','cpp'],patch_artist=True,showmeans=True)
plt.show()
```

```python
#step plot
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y=[6,7,8,9,10]
plt.step(x,y,color='orange',marker='*')
plt.grid()
plt.show()
```

```python
#line graph
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y=[5,7,1,9,12]
plt.plot(x,y,color='orange')
plt.grid()
plt.annotate('py',xy=(3,3),xytext=(4,4),arrowprops=dict(facecolor='black'))
plt.text(2,10,'py',fontsize=17)
plt.show()
```

```python
#stem-diagram
x=[1,2,3,4]
y=[6,7,8,9]
plt.stem(x,y,linefmt=":",markerfmt="b*")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([2, 3, 4, 5])
y = np.array([4, 9, 16, 25])

# Plot line
plt.plot(x, y, color='red', linewidth=5)

# Fill area between curve and x-axis
plt.fill_between(x, y, color='pink', where=(x >= 2) & (y <= 16))

# Display the plot
plt.show()
```
